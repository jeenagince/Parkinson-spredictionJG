```python
#4) knn using loo cross validation
import pandas as pd
from sklearn.model_selection import LeaveOneOut
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Apply LOO method
LOO_KNN_EX = LeaveOneOut()

# Initialize lists to store true and predicted labels
y_true_exknn, y_pred_exknn = [], []

for train_index, test_index in LOO_KNN_EX.split(X_ex):
    # Split data into training and testing sets based on LOO indices
    X_train_ex, X_test_ex = X_ex.iloc[train_index], X_ex.iloc[test_index]
    y_train_ex, y_test_ex = y_ex.iloc[train_index], y_ex.iloc[test_index]

    # Build and train the KNN model

    knn_classifier_cv = KNeighborsClassifier(n_neighbors=5)
    knn_classifier_cv.fit(X_train_ex, y_train_ex)

    # Make predictions
    y_pred_exknn.append(knn_classifier_cv.predict(X_test_ex)[0])
    y_true_exknn.append(y_test_ex.values[0])  # Convert y_test to a single value

# Calculate performance of model using the LOO predictions and true labels
KNNCV_EX_ACC1 = accuracy_score(y_true_exknn, y_pred_exknn)
print("Accuracy using LOO and KNN:", KNNCV_EX_ACC1)
print("Precision Score using LOO and KNN: ",precision_score(y_true_exknn, y_pred_exknn,

                                           average='weighted'))
print("Recall Score using LOO and KNN : ",recall_score(y_true_exknn, y_pred_exknn,

                                           average='weighted'))
print("f1 Score using LOO and KNN : ",f1_score(y_true_exknn, y_pred_exknn,

                                           average='weighted'))
```

```python
#3)RANDOM FOREST-USING CROSS VALIDATION TECHNIQUE
from sklearn.datasets import make_blobs
from sklearn.model_selection import LeaveOneOut
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# create loocv procedure
LOO_RF_EX = LeaveOneOut()

# enumerate splits
y_true_rf_ex, y_pred_rf_ex = list(), list()
for train_ix, test_ix in LOO_RF_EX.split(X_ex):
    # split data
    X_train_ex, X_test_ex = X_ex.iloc[train_ix], X_ex.iloc[test_ix]
    y_train_ex, y_test_ex = y_ex.iloc[train_ix], y_ex.iloc[test_ix]

    # fit model
    rf_classifier_cv = RandomForestClassifier(random_state=1)
    rf_classifier_cv .fit(X_train_ex, y_train_ex)

    # store
    y_true_rf_ex.append(y_test_ex.values[0])
    y_pred_rf_ex.append(rf_classifier_cv.predict(X_test_ex)[0])

# calculate performance of model
RFCV_EX_ACC1= accuracy_score(y_true_rf_ex, y_pred_rf_ex)
print('Accuracy: ', RFCV_EX_ACC1)
print("Precision Score : ",precision_score(y_true_rf_ex, y_pred_rf_ex,
                                           average='weighted'))
print("Recall Score : ",recall_score(y_true_rf_ex, y_pred_rf_ex,
                                           average='weighted'))
print("f1 Score : ",f1_score(y_true_rf_ex, y_pred_rf_ex,
                                           average='weighted'))
```
